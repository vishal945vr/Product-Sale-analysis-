```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("book1.xlsx")
print(df.head(5))
```

```python
df.describe()
```

```python
df.drop_duplicates().sum()
```

```python
df.dropna()
print(df.head(5))
```

```python
df.drop(["Status","unnamed1" ],axis=1,inplace=True)
```

```python
df.isnull()
```

```python
df.shape
```

#graphs


#zone


```python
sns.set(rc={"figure.figsize":(5,5)})
sns.countplot(x="Zone",data=df,hue="Gender")
plt.show()
```

#Occuption


```python
sns.set(rc={"figure.figsize":(25,5)})
sns.histplot(x="Occupation",data=df,kde=True)
plt.show()
```

#Age


```python
sns.set(rc={"figure.figsize":(5,5)})
sns.barplot(y="Age Group",x="Gender",data=df,hue="Marital_Status",orient="h")
plt.show()
```

```python
sns.set(rc={"figure.figsize":(15,5)})
sns.lineplot(y="Amount",x="Occupation",data=df,legend=True)
plt.show()
```

```python
sns.set(rc={"figure.figsize":(5,5)})
vr=sns.countplot(x="Gender",data=df,color="r")
for bars in vr.containers:
    vr.bar_label(bars)
```

```python
ax=sns.countplot(data=df,x="Age Group",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
vx=sns.scatterplot(data=df,x="Age Group",y="Amount",hue="Gender")
for bars in ax.containers:
    vx.bar_label(bars)
```

#state


```python
sns.set(rc={"figure.figsize":(25,5)})
st=sns.barplot(x="State",y="Amount",data=df,color="y")
for bars in st.containers:
    st.bar_label(bars)
```

```python
sns.set(rc={"figure.figsize":(25,5)})
syt=sns.barplot(x="State",y="Occupation",data=df,)
for bars in syt.containers:
    syt.bar_label(bars)
```

```python
sns.violinplot(x="State",y="Marital_Status",data=df)
```

#Marital_Status

```python
sns.set(rc={"figure.figsize":(5,5)})
fr=sns.barplot(x="Marital_Status",y="Amount",data=df,hue="Gender")
for bars in fr.containers:
    fr.bar_label(bars)
```

#Product Category

```python
sns.set(rc={"figure.figsize":(25,5)})
sns.histplot(x="Product_Category",data=df)
```

```python
sns.set(rc={"figure.figsize":(25,5)})
sns.barplot(x="Product_Category",y="Amount",data=df)
```
